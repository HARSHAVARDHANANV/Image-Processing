```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.ones((500, 500, 3), dtype="uint8") * 255

plt.imshow(img)
plt.title("Blank Image")
plt.axis("off")
plt.show()
```

```python
line_img = img.copy()

cv2.line(line_img, (50, 50), (450, 50), (255, 0, 0), 5)

plt.imshow(line_img)
plt.title("Line")
plt.axis("off")
plt.show()
```

```python
rect_img = img.copy()

cv2.rectangle(rect_img, (100, 100), (400, 300), (0, 255, 0), 4)

plt.imshow(rect_img)
plt.title("Rectangle")
plt.axis("off")
plt.show()
```

```python
circle_img = img.copy()

cv2.circle(circle_img, (250, 250), 100, (255, 0, 255), 5)

plt.imshow(circle_img)
plt.title("Circle")
plt.axis("off")
plt.show()
```

```python
text_img = img.copy()

cv2.putText(
    text_img,
    "IPCV LAB",
    (120, 250),
    cv2.FONT_HERSHEY_SIMPLEX,
    1.5,
    (0, 0, 255),
    4
)

plt.imshow(text_img)
plt.title("Text")
plt.axis("off")
plt.show()
```

```python
final_img = img.copy()

cv2.line(final_img, (50, 50), (450, 50), (255, 0, 0), 5)

cv2.rectangle(final_img, (70, 100), (220, 250), (0, 255, 0), 4)

cv2.circle(final_img, (350, 180), 70, (255, 0, 255), 5)

cv2.putText(
    final_img,
    "OpenCV",
    (140, 400),
    cv2.FONT_HERSHEY_SIMPLEX,
    1.2,
    (0, 0, 255),
    3
)

plt.imshow(final_img)
plt.title("Geometric Primitives")
plt.axis("off")
plt.show()
```
