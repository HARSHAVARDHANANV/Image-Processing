```python
import cv2
import matplotlib.pyplot as plt
```

```python
main = cv2.imread("main.jpg")

template = cv2.imread("template.jpg")

main_rgb = cv2.cvtColor(main, cv2.COLOR_BGR2RGB)

template_rgb = cv2.cvtColor(template, cv2.COLOR_BGR2RGB)

plt.imshow(main_rgb)
plt.title("Main Image")
plt.axis("off")
plt.show()

plt.imshow(template_rgb)
plt.title("Template Image")
plt.axis("off")
plt.show()
```

```python
main_gray = cv2.cvtColor(main, cv2.COLOR_BGR2GRAY)

template_gray = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
```

```python
result = cv2.matchTemplate(
    main_gray,
    template_gray,
    cv2.TM_CCOEFF_NORMED
)

min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

print("Matching Value:", max_val)
```

```python
h, w = template_gray.shape

top_left = max_loc

bottom_right = (top_left[0] + w, top_left[1] + h)

output = main_rgb.copy()

cv2.rectangle(output, top_left, bottom_right, (255, 0, 0), 4)

plt.imshow(output)
plt.title("Detected Object")
plt.axis("off")
plt.show()
```
