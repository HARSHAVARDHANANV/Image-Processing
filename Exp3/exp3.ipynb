```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("sample.jpg")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
blur = cv2.blur(img_rgb, (10, 10))

plt.imshow(blur)
plt.title("Average Blur")
plt.axis("off")
plt.show()
```

```python
gaussian = cv2.GaussianBlur(img_rgb, (11, 11), 0)

plt.imshow(gaussian)
plt.title("Gaussian Blur")
plt.axis("off")
plt.show()
```

```python
median = cv2.medianBlur(img_rgb, 7)

plt.imshow(median)
plt.title("Median Blur")
plt.axis("off")
plt.show()
```

```python
median = cv2.medianBlur(img_rgb, 7)

plt.imshow(median)
plt.title("Median Blur")
plt.axis("off")
plt.show()
```

```python
kernel = [[0, -1, 0],
          [-1, 5,-1],
          [0, -1, 0]]

import numpy as np

kernel = np.array(kernel)

sharpen = cv2.filter2D(img_rgb, -1, kernel)

plt.imshow(sharpen)
plt.title("Sharpened Image")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

edges = cv2.Canny(gray, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.axis("off")
plt.show()
```
