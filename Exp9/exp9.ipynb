```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("sample.jpg")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
rows, cols = img_rgb.shape[:2]

matrix = np.float32([[1, 0, 100],
                     [0, 1, 50]])

translated = cv2.warpAffine(img_rgb, matrix, (cols, rows))

plt.imshow(translated)
plt.title("Translated Image")
plt.axis("off")
plt.show()
```

```python
scaled = cv2.resize(img_rgb, None, fx=1.5, fy=1.5)

plt.imshow(scaled)
plt.title("Scaled Image")
plt.axis("off")
plt.show()
```

```python
center = (cols // 2, rows // 2)

rotate_matrix = cv2.getRotationMatrix2D(center, 45, 1)

rotated = cv2.warpAffine(img_rgb, rotate_matrix, (cols, rows))

plt.imshow(rotated)
plt.title("Rotated Image")
plt.axis("off")
plt.show()
```

```python
flipped = cv2.flip(img_rgb, 1)

plt.imshow(flipped)
plt.title("Flipped Image")
plt.axis("off")
plt.show()
```

```python
pts1 = np.float32([[50,50], [200,50], [50,200]])

pts2 = np.float32([[10,100], [200,50], [100,250]])

affine_matrix = cv2.getAffineTransform(pts1, pts2)

affine = cv2.warpAffine(img_rgb, affine_matrix, (cols, rows))

plt.imshow(affine)
plt.title("Affine Transformation")
plt.axis("off")
plt.show()
```
