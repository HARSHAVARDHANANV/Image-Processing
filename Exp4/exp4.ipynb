```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('sample.jpg')
```

```python
img = cv2.imread("sample.jpg")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Gray Image")
plt.axis("off")
plt.show()
```

```python
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(binary, cmap='gray')
plt.title("Binary Threshold")
plt.axis("off")
plt.show()
```

```python
_, inverse = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

plt.imshow(inverse, cmap='gray')
plt.title("Binary Inverse")
plt.axis("off")
plt.show()
```

```python
_, trunc = cv2.threshold(gray, 127, 255, cv2.THRESH_TRUNC)

plt.imshow(trunc, cmap='gray')
plt.title("Trunc Threshold")
plt.axis("off")
plt.show()
```

```python
adaptive = cv2.adaptiveThreshold(
    gray,
    255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11,
    2
)

plt.imshow(adaptive, cmap='gray')
plt.title("Adaptive Threshold")
plt.axis("off")
plt.show()
```

```python
_, otsu = cv2.threshold(
    gray,
    0,
    255,
    cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

plt.imshow(otsu, cmap='gray')
plt.title("Otsu Threshold")
plt.axis("off")
plt.show()
```
